## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Our dataset is all clean, it has zero null values
df.info()
````

````python
print(" Missing Values what")
missing = df.isnull().sum()
if missing.sum()>0:
   print("Missing values are there")
else:
   print("No missing values found")

print("\n Duplicate Rows ")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

print("\n Outlier Detection (IQR method for numeric columns) ")                      # <--- IQR - Interquirtile Range
numeric_cols = df.select_dtypes(include='number').columns                            # <--- selects all numeric columns
for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]   # <--- gives the values inbetween our IQ Range
        print(f"{col}: {outliers.shape[0]} outliers")

        sns.boxplot(x=df[col])
        plt.title(f"Outliers in {col}")
        plt.show()
````

If you see the charts, there are outliers according to the formula, but we'll never know whether they are actual outliers or genuine data, so for now lets keep them as they are

````python
# Credit History Age this column has values like (265, 266, 267) ,it seems like these are months lets convert them to years
df['Credit_History_Age_Years'] = df['Credit_History_Age'] / 12
````

````python
# categorisation of the delay on basis of delay

def delay_classification(days):

        if days <= 0:
            return 'On-Time'
        elif days < 15:
            return 'Slightly Late'
        else :
            return 'Very Late'
````

````python
df['Delay_Category'] = df['Delay_from_due_date'].apply(delay_classification) # <--- New column is made with categorisation
````

````python
# Debt to Income ratio column
df['Debt_to_Income'] = df['Outstanding_Debt'] / df['Annual_Income']
````

````python
df['EMI_to_Salary'] = df['Total_EMI_per_month'] / df['Monthly_Inhand_Salary']
````

````python
df['EMI_to_Salary']
````

````python
def age_group(age):
  if age < 20:
    return 'Under 20'
  elif age < 30:
    return "Under 30"
  elif age < 40:
    return 'Under 40'
  elif age < 50:
    return 'Under 50'
  elif age < 60:
    return 'Under 60'
  else:
    return 'Senior Citizen'

df['Age_Group'] = df['Age'].apply(age_group)
````
